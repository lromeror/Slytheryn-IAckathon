```python
import pandas as pd
import numpy  as np
```

```python
planning = pd.read_csv("../../../planning2015-2023.csv")
tender = pd.read_csv("../../../tender2015-2023.csv")
releases  = pd.read_csv("../../../releases2015-2023.csv")
contracts  = pd.read_csv("../../../contracts2015-2023.csv")
```

```python
planning.info(),tender.info(),releases.info(),contracts.info()
```

```python
# Indexa los dataframes por 'ocid' para acelerar las operaciones de fusión
contracts.set_index('ocid', inplace=True)
planning.set_index('ocid', inplace=True)
releases.set_index('ocid', inplace=True)
tender.set_index('ocid', inplace=True)

# Renombra las columnas una vez fuera del bucle
planning.columns = planning.columns + '_planning'
releases.columns = releases.columns + '_releases'
tender.columns = tender.columns + '_tender'

# Realiza las fusiones usando merge
df_global = contracts.merge(planning[["budget_description_planning","budget_amount_planning","budget_currency_planning","año_planning"]], left_index=True, right_index=True, how='left')
df_global = df_global.merge(releases[["buyer_name_releases","año_releases"]], left_index=True, right_index=True, how='left')
df_global = df_global.merge(tender[['description_tender', 'status_tender', 'procuringEntity_id_tender', 'procuringEntity_name_tender', 'value_amount_tender','value_currency_tender', 'procurementMethod_tender', 'procurementMethodDetails_tender','mainProcurementCategory_tender','año_tender']], left_index=True, right_index=True, how='left')

# Restablece el índice
df_global.reset_index(inplace=True)
#df_global.to_csv("../../../df_global.csv")
```

```python
df_global = pd.read_csv("../../../df_global.csv")
```
