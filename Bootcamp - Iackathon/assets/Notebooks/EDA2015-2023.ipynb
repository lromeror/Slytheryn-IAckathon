De los 1787562 contratos que tienen amount null 1587262 estan terminados y 200300 estan active

```python
#saber a que años pertenecen los contratos activos con amount null
año_contract_amount_null = contracts[(contracts['amount'].isnull()) & (contracts['status']=='active')]['año'].value_counts()
año_contract_amount_null = pd.DataFrame(año_contract_amount_null)
año_contract_amount_null
```
